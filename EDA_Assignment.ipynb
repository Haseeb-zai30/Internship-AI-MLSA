#### Deal with Missing values

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
  df[col].fillna(df[col].mean(), inplace=True)
```

```python
for col in df.select_dtypes(include=['object']).columns:
  df[col].fillna(df[col].mode()[0], inplace=True)
```

#### Encoding

**Identify columns and then find a suitable technique for it**

## Nominal Encoding

```python
nominal_cols = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'MaritalStatus', 'Over18', 'OverTime']
df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
```

## Ordinal Encoding

```python
possible_ordinal_cols = [
    'Education',
    'EnvironmentSatisfaction',
    'JobSatisfaction',
    'PerformanceRating',
    'WorkLifeBalance',
    'RelationshipSatisfaction'
]

# Keep only columns that are in the dataset
ordinal_cols = [col for col in possible_ordinal_cols if col in df.columns]

encoder = OrdinalEncoder()
df[ordinal_cols] = encoder.fit_transform(df[ordinal_cols])
```
